# **Checking Salary is less than 100k or more using decision tree**

```python
import pandas as pd
```

```python
df=pd.read_csv('salaries.csv')
df
```

```python
x=df[['company','job','degree']]
x
```

```python
y=df.salary_more_then_100k
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lecompany=LabelEncoder()
lejob=LabelEncoder()
ledegree=LabelEncoder()
```

```python
df['company_n']=lecompany.fit_transform(df['company'])
df['job_n']=lejob.fit_transform(df['job'])
df['degree_n']=ledegree.fit_transform(df['degree'])
```

```python
df
```

```python
x=df.drop(['company','job','degree','salary_more_then_100k'],axis='columns')
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[2,2,1]])
```

# **checking passenger is survived or not using decision tree**

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df
```

```python
lesex=LabelEncoder()
df['sex_n']=lesex.fit_transform(df['Sex'])
df
```

```python
df=df.drop(['Sex'],axis='columns')
df
```

```python
x=df.drop(['Survived'],axis='columns')
x
```

```python
y=df.Survived
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```

```python
model.predict([[3,22.0,7.2500,1]])
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,model.predict(x_test))
cm
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
